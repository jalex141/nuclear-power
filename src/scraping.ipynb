```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.scielo.br/j/motriz/a/xjNfHRWhzpxWMDvYjgKQKGP/?lang=en#ModalTablet2"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content)
```

```python
body = soup.find_all(name="div",
    class_="modal-body"
).
```

```python
cols = body[3].find_all("tr")
cols= cols[2:]
```

```python
def get_stat(col):
    lines = col.find_all("td")
    
    category = lines[0].text
    lw_male_m= lines[1].text
    lw_male_sdev = lines[2].text
    mw_male_m= lines[3].text
    mw_male_sdev= lines[4].text
    hw_male_m= lines[5].text
    hw_male_sdev= lines[6].text
    lw_female_m= lines[7].text
    lw_female_sdev= lines[8].text
    mhw_female_m= lines[9].text
    mhw_female_sdev= lines[10].text

    
    return {"category": category, "lw_male_m": lw_male_m, "lw_male_sdev": lw_male_sdev,
           "mw_male_m": mw_male_m, "mw_male_sdev": mw_male_sdev, "hw_male_m": hw_male_m,
            "hw_male_sdev": hw_male_sdev, "lw_female_m": lw_female_m, "lw_female_sdev": lw_female_sdev,
            "mhw_female_m": mhw_female_m, "mhw_female_sdev": mhw_female_sdev}
```

```python
df = pd.DataFrame([get_stat(col) for col in cols])
```
